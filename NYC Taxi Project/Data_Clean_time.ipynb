```python
#extract a column (time, in the type of str)
temp_g_time = list()
#temp['tpep_pickup_datetime']
for y in temp_g['dropoff_time']:
    temp_g_time.append(y)
```

```python
#extract another list
temp_g_time = list()
#temp['tpep_pickup_datetime']
for y in temp_g['pickup_time']:
    temp_g_time.append(y)
```

```python
#convert time data from str to float, making it easy for filtering, we didn't use 'datetime' package since it takes so long
#result is time series in the type of float
result = []
for item in temp_g_time:
    pattern = '^.+:'
    regex = re.compile(pattern)
    y = regex.findall(item)
    x = float(y[0][:-1])
    pattern_1 = ':.+$'
    regex_1 = re.compile(pattern_1)
    yy = regex_1.findall(item)
    xx = float(yy[0][1:])
    #temp_g['1'][count] = x + xx/100
    result.append(x + xx / 100)
result
```

```python
#filter by time, return the list of index
#index_list is a list of numbers that contains the indexes of rows that meet certain requirment
index_list = []
count = 0
for x in result :
    if x >= 17.00:
        index_list.append(count)
    count = count + 1
index_list
```

```python
#same as above, different requirment
index_list = []
count = 0
for x in result :
    if x <= 12.00:
        index_list.append(count)
    count = count + 1
index_list
```

```python
#grab rows that meet requirment
temp_g = temp_g.loc[index_list]
temp_g
```
